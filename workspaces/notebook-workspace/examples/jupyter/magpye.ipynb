## Summer of WeatherCode 2022: Bringing Magics weather maps to Matplotlib



## Possible approach
We use the binary format of Magics, and reload it in matplotlib

```python
from magpye import GeoMap

fig = GeoMap(area_name='europe')

fig.coastlines(land_colour="cream",resolution="medium")
```

**t850.grib** contains  a temperature field at 850 hPa.   We use a shading going from the blue for cold temperature to red for hot   
In the show method , we ask Magics to generate a png that we display in an ipython image. 

```python
fig.contour_shaded("t850.grib", style="rainbow_temperature1")

fig.coastlines(resolution="medium")
fig.gridlines()
fig.show()
```

### Trying now to use matpltolib
We are going to save the figure using the internal magics binary format. 

```python
mgb = "magics-binary.mgb"
fig.save(mgb)
print("the figure has been saved in a file called magics-binary.mgb using the magics binary format")
```

### We now read and interpret the binary information. 


```python
from Magics import binary
%matplotlib widget
```

```python
binary.plot_mgb(mgb)
```
